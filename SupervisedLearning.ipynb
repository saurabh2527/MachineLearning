# Supervised Learning.

### Regression.
- It predict a continuous numerucal value based on  input data.

### Linear Regression.
- It predict continuous value by finding a straight line  relationship between input and output 

 - 1. Linear Regression.
    - Finds a Pattern in Old Data.
    - Straight Line
    - Line
    - y=mx+b

```python
from sklearn.linear_model import LinearRegression
```

```python
X=[[1],[2],[3],[4],[5]]
y=[40,50,60,70,80]

model= LinearRegression()

model.fit(X,y)

hours=float(input("Enter the number of hours studies="))

predicted_marks=model.predict([[hours]]) # [[]] It is a 2D List 


print(f"Based on the number of hours studies {hours}, the predicted marks are {predicted_marks}")
```
