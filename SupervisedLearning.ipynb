# Supervised Learning.
 - Supervised learning is a machine learning approach where the model is trained using labeled data, meaning each    data point has:
   - ✔ Input features (X)
   - ✔ Correct output / label (Y)
   - The algorithm learns the relationship between inputs and outputs to make predictions on new data. 

  ### Regression.
    - It predict a continuous numerical value based on  input data.

  ### Linear Regression.
    - It predict continuous value by finding a straight line  relationship between input and output 

   - 1. Linear Regression.
    - Finds a Pattern in Old Data.
    - Straight Line
    - Line
    - y=mx+b

```python
from sklearn.linear_model import LinearRegression
```

```python
X=[[1],[2],[3],[4],[5]]
y=[40,50,60,70,80]

model= LinearRegression()

model.fit(X,y)

hours=float(input("Enter the number of hours studies="))

predicted_marks=model.predict([[hours]])[0] # [[]] It is a 2D List 


print(f"Based on the number of hours studies {hours}, the predicted marks are {predicted_marks}")
```

# Classification.
  - Classification is a type of supervised learning where the model learns from labeled data and predicts a category (class) for new data

  ### Logistic Regression.
      - Logistic Regression is a supervised machine learning algorithm that uses the sigmoid function to predict the probability of an input belonging to a particular class, mainly for binary classification.

```python
from sklearn.linear_model import LogisticRegression
```

```python
X=[[1],[2],[3],[4],[5]]
y=[0,0,1,1,1]

model= LogisticRegression()

model.fit(X,y)

hours=float(input("Enter the number of hours studies="))

result=model.predict([[hours]])[0] # [[]] It is a 2D List

if result==1:
    print(f"Based on the number of hours studies {hours}, the predicted result is Pass")

else:
    print(f"Based on the number of hours studies {hours}, the predicted result is Fail")


```

### K-Nearest Neigbhour.
   - K-Nearest Neighbour (KNN) is a supervised machine learning algorithm used for classification and regression.
   - It classifies a data point based on the majority class of its K nearest neighbors.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = [
    [180,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5]
]

# 0=Apple , 1=Orange 
y=[0,0,0,1,1,1]

model= KNeighborsClassifier(n_neighbors=3) #

model.fit(X,y)

weight=float(input("Enter a weight in grams:"))
size=float(input("Enter a size in Cm:"))

prediction = model.predict([[weight,size]])[0] #

if prediction==0:
    print("This is an apple:")

else:
    print("This is an orange:")
```

# Dession Tree.
  - A Decision Tree is a supervised machine learning algorithm used for classification and regression that makes predictions by asking yes/no questions and splitting data into a tree-like structure, with final results at the leaf nodes.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X=[
    [7,2],
    [8,3],
    [9,8],
    [10,9]
]

y=[0,0,1,1]

model= DecisionTreeClassifier()

model.fit(X,y)

size=float(input("Enter a size in Cm:"))
shade=float(input("Enter a shade in scale of 1-10:"))

result=model.predict([[size,shade]])[0]

if result==0:
    print("This is likly an apple:")

else:
    print("This is an orange:")
```

### .fit()
   - The .fit() method is used to train the machine learning model.

### .Predict()
   - The .predict() method is used to make predictions on new or unseen data after the model has been trained.
