```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,root_mean_squared_error
import numpy as np
```

```python
df = pd.DataFrame({
    "hours":[1,2,3,4,5,6],
    "scores":[52,57,65,70,75,80]
})
```

```python
print(df.head())
```

```python
X = df[["hours"]] # double brackets to make it a 2D array
y = df["scores"] # Traget column

model = LinearRegression()
model.fit(X,y)

predicted_scores = model.predict(X)

# Evaluation
mae = mean_absolute_error(y, predicted_scores)
mse = mean_squared_error(y, predicted_scores)
rmse = np.sqrt(mse)

# show the results

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)

new_hours=float(input("Enter the number of hours studied: "))
new_prediction = model.predict([[new_hours]])
print(f"Predicted score for {new_hours} hours of study is:", new_prediction[0])
```
