```python
data=pd.read_csv("Cybersecurity Incidents and Data Breaches 2024.csv")
print(data.head())
```

```python
data_label = data.copy() # Creating a copy of the original dataframe
le = LabelEncoder() # Initializing LabelEncoder
data_label["MFA_Enabled_Encoded"] = le.fit_transform(data_label["MFA_Enabled"]) # fit_transform means to fit the encoder and transform the data in one step
print('\nLabel Encoded data')
print(data_label[["MFA_Enabled", "MFA_Enabled_Encoded"]].head(10))
```

```python
df_encoded = pd.get_dummies(data, columns=["Industry"]) # get_dummies means converting categorical variable into binary 

print("\nOne-Hot Encoded Data")
print(df_encoded.head())
```
